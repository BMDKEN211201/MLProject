#### Since the median and mode are same in all the cases we directly impute the mean in the place of 0s ####

```python
# prompt: Since the median and mode are same in all the cases we directly impute the mean in the place of 0s

data[['x','y','z']] = data[['x','y','z']].replace(0,data[['x','y','z']].mean())
```

### KDE for Numerical Variables ###

```python
# 1. Create a list of numerical features
num_features = data.select_dtypes(include=['float64', 'int64']).columns.tolist()
# 2. Loop through num_features and then plot each feature
print(num_features)
```

#### Apply the log transformation ####

```python
# Loop through num_features and apply np.log function
for feature in num_features:
    data[feature] = np.log(data[feature])
```

```python
# Check some rows
data.head(20)
```

### Handling Categorical Variables ###

```python
# Find the value counts for the categorical variables
categorical_feat=['cut','color','clarity']
# Looping
for feature in categorical_feat:
    print(data[feature].value_counts())
```

#### Apply Label Encoding ####

```python
# Apply label encoder on all the categorical variables
le = LabelEncoder()
for feature in categorical_feat:
    data[feature] = le.fit_transform(data[feature])
data.head(10)
```

```python
data[feature_scale] = data[feature_scale].replace([np.inf, -np.inf], np.nan)
```

```python
data[feature_scale] = data[feature_scale].fillna(data[feature_scale].mean())
```

### Feature Scaling ###

```python
# Scale all the features except Price
# 1. Create a list of all the features that need to be scaled
feature_scale=[feature for feature in data.columns if feature not in ['price']]
# Import and call
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
# Fit
scaler.fit(data[feature_scale])
data[feature_scale]=scaler.transform(data[feature_scale])
data.head()
```

### Split train and Test Data ###

```python
# y is your target column and x is your feature columns
x = data.drop(['price'],axis=1)
y = data['price']
```

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

### Train the model ###

```python
# Call the model
regressor = LinearRegression()
# Fit
regressor.fit(X_train, y_train)
# Print results
print(regressor.intercept_)
print(regressor.coef_)
```

```python
#### Predcit the results using the created model ####
y_pred=regressor.predict(X_test)
# Create a dataframe of actual values and predicted values of the test set.
data_result=pd.DataFrame({'Actual':y_test,'Predict':y_pred})
```

### Evaluate the model ###

```python
# Please pick some metrics of your interest
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
y_pred = regressor.predict(X_test)
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R-squared:', r2_score(y_test, y_pred))
# Root Mean Squared Error (RMSE)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("Root Mean Squared Error (RMSE):", rmse)



```
