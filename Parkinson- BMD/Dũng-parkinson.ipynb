```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Importing dataset**

```python
df = pd.read_csv("/content/drive/MyDrive/ML/Class 1 - EDA/Datasets/parkinsons_new.csv")
```

**Pre-Processing & EDA**

```python
# Check head
df.head(10)
```

```python
# Check Null values
df.isnull().sum().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
#dropping a particular column 'name'
df.drop(columns=['name'], inplace=True)
```

```python
#  Get dummies for sex and then drop the original 'sex' column
sex_dummies = pd.get_dummies(df['sex'])
df = pd.concat([df, sex_dummies], axis = 1)
df = df.drop(columns='sex')
```

```python
# Split target and features
y = df['status']
X = df.drop(columns='status')
```

**Train-Test split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

**Standardization of Data**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler().fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

**Classification SVM**

**Linear**

```python
# Import SVM model
from sklearn import svm
```

```python
# Create a svm Classifier (kernel = 'linear')
clf = svm.SVC(kernel = "linear") # Linear Kernel

# Train the model using the training sets
clf.fit(X_train, y_train)
```

```python
# Predict Xtrain
train_pred = clf.predict(X_train)
```

```python
# Predict Xtest
test_pred = clf.predict(X_test)
```

**Models Performance**

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score, auc, roc_curve

# Confusion Matrix Of Training Set

cm_train = confusion_matrix(y_train, train_pred)

print('Training Confusion matrix\n\n', cm_train)


# Confusion Matrix Of Test Set

cm_test = confusion_matrix(y_test, test_pred)

print('Test Confusion matrix\n\n', cm_test)
```

```python
# Accuracy of the training set
acc_train_set = accuracy_score(y_train, train_pred)

print(f'The accuracy score of the training set {acc_train_set}')
```

```python
# Accuracy of the test set

acc_test_set = accuracy_score(y_test, test_pred)

print(f'The accuracy score of the test set {acc_test_set}')
```

**Grid Search**

```python
from sklearn.model_selection import GridSearchCV
parameters = {'C': [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50, 100, 500, 1000, 5000],
              'degree': [2, 3, 4, 5],
              'gamma':[0.001, 0.01, 0.1, 0.5, 1],
              'kernel': ['rbf','poly']
              }
```

```python
from sklearn.metrics import classification_report

grid = GridSearchCV(clf, param_grid=parameters)

grid.fit(X_train,y_train) # training set

grid_prediction = grid.predict(X_test) # test set

# Classification report

print(classification_report(y_test, grid_prediction)) # actual vs prediction

acc_grid = accuracy_score(y_test, grid_prediction) # actual vs prediction

print(f'The accuracy score of the of gridsearchcv random forest {acc_grid}')
```

```python
# Random forest
from sklearn.ensemble import RandomForestClassifier

# Create the model

random_forest = RandomForestClassifier(random_state=0)

# Fit the model

random_forest.fit(X_train, y_train)

# Predict the test result

rfc_pred = random_forest.predict(X_test)

rfc_acc = accuracy_score(y_test, rfc_pred)

print(f'The accuracy score of the training set {rfc_acc}')

print(classification_report(y_test, rfc_pred))
```

```python
# Feature importance with random forest and rerun SVM and RF
feature_scores = pd.Series(random_forest.feature_importances_, index=df.drop(columns='status').columns).sort_values(ascending=False)
```

```python
selected_features = feature_scores[feature_scores > 0.05]
```

```python
# Choose variable with feature score > 0.05
# Rerun Random Forest

# Split the data again

y = df['status']
X = df[selected_features.index]

X_train = X_train[selected_features]

X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X, y, test_size=0.2, random_state=2)

sc = StandardScaler().fit(X_train_2)
X_train_2 = sc.transform(X_train_2)
X_test_2 = sc.transform(X_test_2)

random_forest_tuned = RandomForestClassifier()

random_forest_tuned.fit(X_train_2, y_train_2)

# Predict the test result

rfc_pred_tuned = random_forest_tuned.predict(X_test_2)

rfc_acc_tuned = accuracy_score(y_test_2, rfc_pred_tuned)

print(f'The accuracy score of the training set {rfc_acc_tuned}')

print(classification_report(y_test, rfc_pred_tuned))
```

```python
# Choose variable with feature score > 0.05
# Rerun Random Forest

# Select target and features
y = df['status']
selected_features = feature_scores[feature_scores > 0.05].index.tolist()  # Convert to list
X = df[selected_features]  # Use df[selected_features] for feature selection

# Now split the data using train_test_split
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X, y, test_size=0.2, random_state=2)

# Standard scaling
sc = StandardScaler().fit(X_train_2)
X_train_2 = sc.transform(X_train_2)
X_test_2 = sc.transform(X_test_2)

# Train the Random Forest model
random_forest_tuned = RandomForestClassifier()
random_forest_tuned.fit(X_train_2, y_train_2)

# Predict on the test set
rfc_pred_tuned = random_forest_tuned.predict(X_test_2)
rfc_acc_tuned = accuracy_score(y_test_2, rfc_pred_tuned)

# Accuracy score
print(f'The accuracy score of the training set: {rfc_acc_tuned}')

# Print classification report
print(classification_report(y_test_2, rfc_pred_tuned))
```

```python
# Choose variable with feature score > 0.05
# Rerun SVM

# Create a svm Classifier
clf_tuned = svm.SVC(kernel = 'linear') # Linear Kernel

# Train the model using the training sets
clf_tuned.fit(X_train_2, y_train_2)

# Predict
clf_pred_tuned = clf_tuned.predict(X_test_2)

cfc_acc_tuned = accuracy_score(y_test_2, clf_pred_tuned)

print(f'The accuracy score of the training set {cfc_acc_tuned}')

print(classification_report(y_test, clf_pred_tuned))
```
