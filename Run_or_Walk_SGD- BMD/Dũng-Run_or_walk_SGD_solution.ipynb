```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt

from subprocess import check_output
from datetime import time
```

Currently, the dataset contains a single file (run_or_walk_dataset_main.zip) which represents sensor data from accelerometer and gyroscope collected from iPhone 5c in 10 seconds interval and ~5.4/second frequency. This data is represented by following columns (each column contains sensor data for one of the sensor's axes):

*   acceleration_x
*   acceleration_y
*   acceleration_z
*   gyro_x
*   gyro_y
*   gyro_z

The dataset contains activity type represented by "activity" column which reflects following activities:

*   "0": walking
*   "1": running

Additionally, the dataset contains "date", "time" and "username" columns which provide information about the exact date, time and user which collected these measurements.

Please note that data represented in this dataset collected on left wrist only.

```python
df = pd.read_csv("/content/drive/MyDrive/ML/Class 1 - EDA/Datasets/Run_or_walk.csv")
print(df.shape)
df.head()
```

```python
df = df.drop(["date", "time", "username"], axis=1)
df.head()
```

```python
print(df.describe())
```

```python
data = df.values
X = data[:, 1:]  # all rows, no label
y = data[:, 0]  # all rows, label only
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Scale the data to be between -1 and 1
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import SGDClassifier
model = SGDClassifier(loss="hinge", penalty="l2")
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
#np.random.shuffle(data)
from sklearn.model_selection import GridSearchCV

params = {
    "alpha" : [0.0001, 0.001, 0.01, 0.1],
    "penalty" : ["l2", "l1", "none"],
}

model = SGDClassifier(max_iter=1000)
clf = GridSearchCV(model, param_grid=params)
```

```python
#X = data[:, 1:]  # all rows, no label
#y = data[:, 0]  # all rows, label only
clf.fit(X_train, y_train)
print(clf.best_score_)
```

```python
print(clf.best_estimator_)
```

```python
# Fit best estimators on the test set
model = SGDClassifier(alpha=0.001, penalty='l1')
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
