# Import libraries

```python
import warnings
warnings.filterwarnings('ignore')
import os
import pandas as pd
from pandas import DataFrame
import pylab as pl
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
```

# Read the dataset

```python
df = pd.read_csv("/content/drive/MyDrive/ML/ML review/Class 9 - Decision trees/Datasets/Admission_Predict.csv")
```

```python
df.head(5)
```

# Drop column Serial No

```python
df = df.drop('Serial No.', axis = 1)
```

# Split X and y (Chance of admit)

```python
y = df['Chance of Admit ']
X = df.drop('Chance of Admit ', axis=1)
```

# Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

# Scale X_train and X_test using either StandardScaler or MinMaxScaler

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
sc = MinMaxScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Call a decision regressor model (Define any random_state)

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor(random_state=42)
```

# Fit the model

```python
dtr.fit(X_train, y_train)
```

# Predict X_test and check some metrics

```python
from sklearn.metrics import r2_score, mean_squared_error

dtr_pred = dtr.predict(X_test)
r2_DTC = r2_score(y_test, dtr_pred)
mse_DTC = mean_squared_error(y_test, dtr_pred)
print(f"The R2 Score Of Decision Tree is {r2_DTC}, MSE is {mse_DTC}")
```

# Fit KNN regressor and check R2

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()
```

```python
knr.fit(X_train, y_train)
knr_pred = knr.predict(X_test)
```

```python
r2_KNN = r2_score(y_test, knr_pred)
mse_KNN = mean_squared_error(y_test, knr_pred)
print(f"The R2 Score Of KNN is {r2_KNN}, MSE is {mse_KNN}")
print("\n")
```

# Use GridsearchCV with estimator = DecisionTreeRegressor and check some metrics

```python
from sklearn.model_selection import GridSearchCV
param_grid = [{"max_depth":[3,4,5, None], "max_features":[3,4,5,6,7]}]
gs = GridSearchCV(estimator=DecisionTreeRegressor(random_state=123),param_grid = param_grid,cv=10)
gs.fit(X_train ,y_train)
gs.cv_results_['params']
gs.best_estimator_
```

```python
tree = DecisionTreeRegressor(max_depth=3,max_features=4)
tree.fit(X_train ,y_train)
predictions = tree.predict(X_test)
df=pd.DataFrame({'Actual':y_test, 'Predicted':predictions})
```

```python
df.head(10)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,predictions))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
print('r2_score:', metrics.r2_score(y_test,predictions))
```
